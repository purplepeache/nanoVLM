```python
# Utils
def get_longest_consecutive_label_end_index(data_list, target_label):
    """
    Returns the index of the *last* element in the longest consecutive 
    sequence of the target_label. If multiple sequences have the same 
    maximum length, the one that appears later in the list is prioritized 
    (though in the given example, the first one is used).
    """
    max_length = 0
    end_index_of_longest = -1
    current_length = 0
    
    # We iterate through the list to find all consecutive sequences
    for i in range(len(data_list)):
        if data_list[i] == target_label:
            # If the current element matches the target, increase the current length
            current_length += 1
        else:
            # If the current element does not match, the sequence is broken
            if current_length > 0:
                # Check if this completed sequence is the new longest one
                if current_length >= max_length:
                    max_length = current_length
                    # The end index is the index *before* the current loop index (i-1)
                    end_index_of_longest = i - 1
            # Reset the counter for the next potential sequence
            current_length = 0
    
    # After the loop, check the very last sequence if it was the target_label
    if current_length > 0:
        if current_length >= max_length:
            # The end index is the last index of the list (len(data_list) - 1)
            end_index_of_longest = len(data_list) - 1
            
    return end_index_of_longest
```

```python
# Doing the benchmark
SPLIT = "validation"
greedy = True

do_sample = False
n_sample = -1
if do_sample:
    data_idxs = np.random.choice(len(D[SPLIT]), size=n_sample, replace=False)
else:
    data_idxs = range(len(D[SPLIT]))

outputs = {
    "data_idx": [],
    "question": [],
    "prompt": [],
    "response": [],
    "correct_choice": [],
    "correct_choice_alphabet": [],
    "n_drop_predicted": [],
}
for data_idx in tqdm(data_idxs):
    # Fetch the datapoint
    data = D[SPLIT][data_idx]
    question = data["question"]
    image = data["image"]
    choices = data["choices"]
    correct_choice_idx = data["correct_choice_idx"]
    correct_choice = choices[correct_choice_idx]
    correct_choice_alphabet = chr(ord("A") + correct_choice_idx)

    outputs["data_idx"].append(data_idx)
    outputs["question"].append(question)
    outputs["correct_choice"].append(correct_choice)
    outputs["correct_choice_alphabet"].append(correct_choice_alphabet)

    # Format the prompt
    prompt = f"""
    {question}\n
    """.strip()
    for i, choice in enumerate(choices):
        prompt += f"\n{chr(ord('A') + i)}. {choice}"
    prompt += "\n\n(You must answer with only the letter enclosed in curly bracket, e.g. {A})"

    outputs["prompt"].append(prompt)

    # Predict n_drop
    n_drop_candidates = [0,2,4,6,8,10]
    correctness_predictions = n_drop_predictor.predict(
        np.array([
            data["question_embedding"] + [i]
            for i in n_drop_candidates
        ])
    ).tolist()
    correctness_predictions = correctness_predictions
    n_drop_predicted = get_longest_consecutive_label_end_index(correctness_predictions, "correct")
    n_drop_predicted = n_drop_candidates[n_drop_predicted]

    # Adaptively drop layers
    model.model.decoder.blocks = decoder_blocks[:(n_decoder_blocks - n_drop_predicted)]
    # print(f"Kept layers: {len(model.model.decoder.blocks)} / {n_decoder_blocks} layers")
    outputs["n_drop_predicted"].append(n_drop_predicted)

    # Generate the response
    response = model.generate_from_image_and_prompt(image=image, prompt=prompt, max_new_tokens=max_new_tokens, greedy=greedy)
    
    outputs["response"].append(response)
    # break
```

```python
def get_enclosed_letter(text: str):
    """
    Finds the first single alphabetical letter that is enclosed by non-alphabetical
    characters or the start/end of the string.

    Args:
        text: The input string to search.

    Returns:
        The single alphabetical letter (str) if found, otherwise None.
    """
    # re.search looks for the pattern anywhere in the string.
    pattern = r"(?:^|[^a-zA-Z])([a-zA-Z])(?:$|[^a-zA-Z])"
    match = re.search(pattern, text)

    if match:
        # The letter is in capturing Group 1 of the regex.
        return match.group(1)
    return None
```

```python
# Parse responses
outputs["parsed_response"] = []
for response_idx, response in enumerate(outputs["response"]):
    if (len(response) == 1) and (response.isalpha()):
        outputs["parsed_response"].append(response.capitalize())
    else:
        parsed_response = get_enclosed_letter(response)
        parsed_response = parsed_response.capitalize() if parsed_response is not None else "Z"
        outputs["parsed_response"].append(parsed_response)

outputs["parsed_response"][:4]
```
